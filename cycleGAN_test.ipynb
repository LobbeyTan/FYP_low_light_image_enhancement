```python
import torch
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from torch.utils.data import DataLoader
from models.cycleGAN import CycleGANModel
from data.custom_image_dataset import CustomImageDataset
from configs.option import Option
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python

model_1 = CycleGANModel(device=device)
model_2 = CycleGANModel(device=device)

model_1.load_model("./checkpoints/training_7/iter_5000")
model_2.load_model("./checkpoints/training_7/iter_15000")
```

```python
dataset = CustomImageDataset(
    img_dir="./datasets/horse2zebra",
    opt=Option(phase="test")
)

batch_size = 3
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
data = next(iter(dataloader))
```

```python
def showImage(paths, row=1, col=1):
    plt.subplots(row, col, figsize=(10,10))

    for i in range(row):
        for j in range(col):
            pos = col * i + j
            idx = pos % len(paths)
            path, name = paths[idx]
            
            plt.subplot(row, col, pos+1)
            plt.imshow(mpimg.imread(path))
            plt.title(name, color="white")
            plt.axis('off')

    plt.show()
    print("-" * 80)
```

```python
for i in range(batch_size):
    showImage([(data["path_A"][i], "Image X"), (data["path_B"][i], "Image Y")], row=1, col=2)
```

```python
paths = model_1.test(data['img_A'], data['img_B'], save_dir="./results/cycleGAN/test_1/")
```

```python
n_images = 6

for i in range(batch_size):
    showImage(paths[i * n_images: (i+1) * n_images], 3, 2)
```

```python
paths = model_2.test(data['img_A'], data['img_B'], save_dir="./results/cycleGAN/test_2")
```

```python
n_images = 6

for i in range(batch_size):
    showImage(paths[i * n_images: (i+1) * n_images], 3, 2)
```
